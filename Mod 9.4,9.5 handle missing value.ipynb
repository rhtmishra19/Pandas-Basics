```python
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv("Datasets/weather_data-Copy1.csv",parse_dates=['day'])
print(df.temperature[0])
type(df.temperature[0])
```

```python
df
```

```python
df.set_index("day",inplace = True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
        "temperature" : df.temperature.mean(),
        "windspeed" : df.windspeed.median(),
        "event" : "No event"
})
new_df
```

```python
new_df = df.fillna(method = "ffill")
new_df
```

```python
new_df = df.fillna(method = "bfill",limit = 1)
new_df
```

```python
new_df = df.fillna(method = "bfill", axis="columns")
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how = "all")
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```

```python
df = df.replace(np.nan, value= -9999)
df
```

```python
new_df = df.replace(to_replace=[-9999.0,40],value=0)
new_df
```

```python
new_df = df.replace({
        "teperature" : -9999,
        "windspeed" : -9999,
        "event" : "Sunny"
},np.nan)
new_df
```
